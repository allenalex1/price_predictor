```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import sklearn
```

```python
# Loading the dataset into a DataFrame

df_house = pd.read_csv('/Users/allenalex/Desktop/AI-ML practice/price_predictor/train.csv', header=0)
df_house[:20]
```

```python
# Creating a new column with City

df_house['City'] = df_house['ADDRESS'].str.split(',').apply(lambda x: x[-1].strip())
df_house[:20]
```

```python
# The function that removes the city from 
def remove_lastpart(address):
    parts = address.split(',')
    return ','.join(parts[:-1]).strip()
```

```python
df_house['ADDRESS'] = df_house['ADDRESS'].apply(remove_lastpart)
```

```python
col = df_house.pop('City')
df_house.insert(9,'CITY', col)
```

```python
# getting unique address
len(df_house['ADDRESS'].unique())
```

```python
# since there are 5915 different addresses, its difficult to represent these into categories 
len(df_house['CITY'].unique())
```

```python
# no of unique cities is 256, hence difficult to represent it as one shot vector
df_house['CITY'].value_counts()
```

```python

# calculate the mean price in each city, there are cities which don't have enough instances to better represent the entire city, 
# we are ignoring that for the time being 
mean_values = df_house.groupby('CITY')['TARGET(PRICE_IN_LACS)'].mean().reset_index()
mean_values
```

```python
df_house = df_house.merge(mean_values, on='CITY', how='left')
df_house[:30]
```

```python
def remove_col(df, col:str):
    df.pop(col)
    return df

def get_dummies(df,col:str):
   df = pd.get_dummies(df, columns=[col], drop_first=True)
   return df
   
```

```python
df_house = get_dummies(df_house,'POSTED_BY')
df_house = get_dummies(df_house,'BHK_OR_RK')
df_house = remove_col(df_house, 'ADDRESS')
df_house = remove_col(df_house, 'CITY')

df_house.astype(float)
df_house.head(30)

# removing address and city cause we cant use it 
```

```python
df_house.astype(float)
```

```python
# checking if there are any Nan values 
df_house.isna().any()
```

```python
#Creating the dataset 

X = df_house
y = df_house['TARGET(PRICE_IN_LACS)_x']
```

```python
X = remove_col(X,'TARGET(PRICE_IN_LACS)_x')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
def create_train_test(X,y,frac):
   X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = frac) 
```

```python
create_train_test(X,y,0.2)
```

```python
len(X_train),len(y_train),len(X_test),len(y_test)
```

```python
#Creating a baseline model using linear regression 

lr = LinearRegression()

def model_fitter(model, X_train, y_train):
    model.fit(X_train, y_train)
# fitting the training and testing data to the model
model_fitter(lr, X_train, y_train)
```

```python
# testing the model

def model_test(model, X_test,y_test):
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_pred,y_test)
    mse = mean_squared_error(y_pred,y_test)
    r2 = r2_score(y_pred,y_test)
    print(f'Mean Absolute Error: {mae}')
    print(f'Mean Squared Error: {mse}')
    print(f'R^2 Score: {r2}')

    return y_pred, mae, mse, r2
```

```python
y_pred,mae,mse,r2 = model_test(lr,X_test,y_test)
```

```python
# Since the error values and R^2 scores are high we'll shift to random forest regression 
rfr = RandomForestRegressor(random_state= 13)
model_fitter(rfr,X_train,y_train)
y_pred,mae,mse,r2 = model_test(rfr,X_test,y_test)
```

```python
# finds the ideal number of estimators that provide the performance 
param_grid = {
    'n_estimators' : [50,100,200,300]
}
rfr = RandomForestRegressor(random_state=13)

grid_search = GridSearchCV(estimator= rfr, param_grid= param_grid,cv=5, scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred2,mae2,mse2,r22 = model_test(best_model, X_test,y_test)
```
